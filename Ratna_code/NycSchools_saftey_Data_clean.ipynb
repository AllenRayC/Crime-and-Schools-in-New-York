```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load 
ss_data_1416 = "2010-2016_School_Safety_Report.csv"
sp_data_1416 = "Combined_HS_Performance_Data.csv"
```

```python
# Read School Safety Data File and store into Pandas Data Frames
ss_data_1416 = pd.read_csv(ss_data_1416)
ss_data_1416.head(5)
```

```python
ss_data_clean_1416 = ss_data_1416.loc[:,["School Year","Building Code","DBN","Location Code","Location Name","Major N","Oth N","NoCrim N","Prop N","Vio N","NTA"]]
ss_data_clean_1416 = ss_data_clean_1416[(ss_data_clean_1416["School Year"] == '2013-14')]
                                        
ss_data_clean_1416.fillna(0, inplace=True)
ss_data_clean_1416.head(5)
```

```python
#Remove zero crime records
ss_data_clean_1416 = ss_data_clean_1416[(ss_data_clean_1416["Major N"] + ss_data_clean_1416["Oth N"] + ss_data_clean_1416["NoCrim N"] + ss_data_clean_1416["Prop N"] + ss_data_clean_1416["Vio N"]) > 0.0 ]
ss_data_clean_1416.head(5)
```

```python
ss_renamed_df = ss_data_clean_1416.rename(columns={"Major N": "Major Crime",
                                                       "Oth N": "Other Crime", 
                                                       "NoCrim N": "Non Criminal",
                                                       "Prop N": "Property Crime",
                                                       "Vio N": "Violent Crime"
                                                      })

ss_renamed_df["NTA"] = ss_renamed_df["NTA"].str.strip()

ss_renamed_df.head(5)
```

```python
#School Safety Data that is clean and sorted by NTA
SS_df = ss_renamed_df.sort_values(["School Year", "NTA", "DBN"], ascending=True)
SS_df.head(5)
```

```python
# Export file as a CSV, without the Pandas index, but with the header
ss_renamed_df.to_csv("SchoolSafety_1314.csv", index=False, header=True)
```

```python
### Note
* This module is to clean the School Performance 
```

```python
# Read School Performance Data File and store into Pandas Data Frames
sp_data_1416 = pd.read_csv(sp_data_1416)
sp_data_1416.head(5)
```

```python
#Remove % from the Grad rate and Coll Rate and convert to float 
for index in range(292):
     sp_data_1416.at[index, "graduation_rate_2014"] = float(sp_data_1416["graduation_rate_2014"][index].replace("%", ""))
     sp_data_1416.at[index, "college_career_rate_2014"] = float(sp_data_1416["college_career_rate_2014"][index].replace("%", ""))
    
```

```python
sp_data_1416
```

```python
sp_data_1416["graduation_rate_2014"] = pd.to_numeric(sp_data_1416["graduation_rate_2014"])
sp_data_1416["college_career_rate_2014"] = pd.to_numeric(sp_data_1416["college_career_rate_2014"])
```

```python
type(sp_data_1416["graduation_rate_2014"][1])
```

```python
sp_data_clean_14 = sp_data_1416.loc[:,["DBN","graduation_rate_2014","college_career_rate_2014","NTA"]]
sp_data_clean_14['School Year'] = '2013-14'
sp_data_clean_14_df = sp_data_clean_14.rename(columns={"graduation_rate_2014": "Grad_rate",
                                                         "college_career_rate_2014": "Coll_rate"                                                       
                                                      })
```

```python
#Sort teh data and save output
SP_df = sp_data_clean_14_df.sort_values(["School Year", "NTA", "DBN"], ascending=True)

SP_df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header

SP_df.to_csv("SchoolPerfomance_1314.csv", index=False, header=True)
```
